##### A DataFrame is a 2D table (rows + columns), like Excel, This is the MOST used structure in pandas


 * In real-world projects, we usually load data from files like CSV, Excel, or databases
##### using functions like:
 * pd.read_csv(), pd.read_excel(), pd.read_sql(), pd.read_parquet()

 * But here, instead of loading data, we are creating our own dataset
 * This is called "synthetic data" — fake but realistic data used for learning

np.random.seed(42)    
* Seed() sets the starting point for random number generation 

🎯 Why we use it  
Without seed:
*  Every run → different data   
*  Your output changes → confusion, chaos, mild existential crisis

With seed:    
*  Same data every time   
*  Easier to debug  
*  Easier to teach 

```python
# ── 3.1 Creating DataFrames ───────────────────────────────────────────────────
# Most common in real work: reading from files
# pd.read_csv(), pd.read_excel(), pd.read_sql(), pd.read_parquet()
import numpy as np
import pandas as pd
np.random.seed(42)
n = 200

# Creating the DataFrame
df = pd.DataFrame({

    # Unique order ID for each order
    # range(start, end) generates sequential numbers
    'order_id': range(1001, 1001 + n),

    # Customer IDs (random numbers between 100 and 159)
    'customer_id': np.random.randint(100, 160, n),

    # Product names randomly selected from a list
    'product': np.random.choice(
        ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'], n
    ),

    # Product category (Electronics or Peripherals)
    'category': np.random.choice(
        ['Electronics', 'Peripherals'], n
    ),

    # Quantity of items per order (between 1 and 5)
    'quantity': np.random.randint(1, 6, n),

    # Price of a single unit (random float between 50 and 2000)
    'unit_price': np.random.uniform(50, 2000, n).round(2),#89.0

    # Discount percentage applied to the order
    'discount_pct': np.random.choice([0, 5, 10, 15, 20], n),

    # Region where the order was placed
    'region': np.random.choice(
        ['North', 'South', 'East', 'West'], n
    ),

    # Sales representative handling the order
    'sales_rep': np.random.choice(
        ['Alice', 'Bob', 'Carol', 'David'], n
    ),

    # Order dates starting from Jan 1, 2024
    # Generates sequential dates
    'order_date': pd.date_range('2024-01-01', periods=n, freq='D')[:n] #periods=n means we create 200 datesnd freq='D' means daily frequency
})

# Derived column
df['revenue'] = (df['quantity'] * df['unit_price'] * (1 - df['discount_pct'] / 100)).round(2)
#revenue = total price * discount
#revenue = (quantity)*price*(1-discount)

print(f'Shape: {df.shape}')#(200, 11)
df.head(5) #first 5 row
```

```python
# ── 3.2 Essential Inspection Methods (first thing you do with ANY dataset) ─────
print('=== SHAPE ===')
print(df.shape)

print('\n=== INFO (dtypes + null counts) ===')
df.info()

print('\n=== DESCRIBE (numeric summary) ===')
df.describe().round(2)

```

```python

# Describe categorical columns too
df.describe(include='object')
```

```python
# ── 3.3 Column Selection — 4 patterns you must know 
# Pattern 1: single column → Series
print(type(df['revenue']))


# Pattern 2: list of columns → DataFrame
subset = df[['order_id', 'product', 'revenue']]
print(subset.head(3))


# Pattern 3: select by dtype
numeric_cols = df.select_dtypes(include='number').columns.tolist()
print(f'\nNumeric columns: {numeric_cols}')

# Pattern 4: filter columns by name pattern
id_cols = [c for c in df.columns if '_id' in c]
print(f'ID columns: {id_cols}')
```

```python
# ── 3.4 Row Selection: .loc vs .iloc ─────────────────────────────────────────

# In pandas, we use .loc and .iloc to select rows (and columns)

# .loc  → label-based selection (uses row/column names)
# .iloc → position-based selection (uses index positions like 0,1,2...)


# Select row by index label
print(df.loc[0, 'product'])

# Select multiple rows + columns
print(df.loc[0:4, ['product', 'revenue', 'region']])

# .iloc — positional, works like Python list slicing
print(df.iloc[0:3, 2:5])  # first 3 rows, columns 2-4

```

```python
# ── 3.5 Filtering (Boolean Masking) — the core of analysis ───────────────────
# Single condition
high_value = df[df['revenue'] > 5000]
print(f'High-value orders: {len(high_value)}')


# Multiple conditions — use & | ~ (NOT 'and' / 'or')
north_laptop = df[(df['region'] == 'North') 
& (df['product'] == 'Laptop')]
print(f'North Laptop orders: {len(north_laptop)}')

# .isin() — cleaner than multiple OR conditions
target_products = df[df['product'].isin(['Laptop', 'Phone'])]
print(f'Laptop + Phone orders: {len(target_products)}')

# .query() — SQL-like, very readable for complex filters
result = df.query('revenue > 3000 and region in ["North", "East"] and discount_pct > 0')
print(f'\nFiltered via query(): {len(result)} rows')

```

```python
# ── 3.6 Adding, Renaming, Dropping Columns ───────────────────────────────────

# Create a copy of the original DataFrame
# This is a good practice so we don’t accidentally modify the original data
df2 = df.copy()


# ── Adding a New Column ──────────────────────────────────────────────────────

# Create a new column 'revenue_per_unit'
# Formula: total revenue divided by quantity

df2['revenue_per_unit'] = (
    df2['revenue'] / df2['quantity']
).round(2)

# round(2) → keeps values up to 2 decimal places


# ── Adding Column with Condition (np.where) ──────────────────────────────────

# np.where works like an if-else condition (vectorized, fast)

# Logic:
# If revenue >= 5000 → 'Large'
# Else if revenue >= 1000 → 'Medium'
# Else → 'Small'

df2['deal_size'] = np.where(
    df2['revenue'] >= 5000, 'Large',
    np.where(df2['revenue'] >= 1000, 'Medium', 'Small')
)


# ── Renaming Columns ─────────────────────────────────────────────────────────

# Rename column names for better readability

df2.rename(
    columns={
        'unit_price': 'price',
        'discount_pct': 'discount'
    },
    inplace=True  # modifies the DataFrame directly
)


# ── Dropping Columns ─────────────────────────────────────────────────────────

# Remove unwanted columns
# Here we remove 'category'

df2.drop(columns=['category'], inplace=True)


# ── Preview Result ───────────────────────────────────────────────────────────

# Show selected columns to verify changes
print(df2[['revenue', 'revenue_per_unit', 'deal_size']].head())
```

```python
# ── 3.7 Sorting ───────────────────────────────────────────────────────────────
# Top 5 orders by revenue
top5 = df.nlargest(5, 'revenue')[['order_id', 'product', 'region', 'revenue']]
print('Top 5 Orders by Revenue:')
print(top5.to_string(index=False))


# Sort by multiple columns
df_sorted = df.sort_values(['region', 'revenue'], ascending=[True, False])
print('\nSorted by Region ASC, Revenue DESC:')
print(df_sorted[['region', 'product', 'revenue']].head(8).to_string(index=False))
```

```python
# ── 3.8 Method Chaining — write clean, readable pandas code ──────────────────
# Bad: intermediate variables everywhere
# temp1 = df[df['region'] == 'North']
# temp2 = temp1[['product', 'revenue']]
# temp3 = temp2.groupby('product')['revenue'].sum()
# result = temp3.sort_values(ascending=False)

# Good: method chaining(perform multiple opration in single flow)
result = (
    df
    # 1: filter rows where region is north
    .query("region == 'North'")

    # 2 select requred colm
    [['product', 'revenue']]

    # 3 group by product
    .groupby('product')['revenue']

    #4 cal total revenue for each prod
    .sum()

    # sort in desc
    .sort_values(ascending=False)

    # 5 convert result back to df
    .reset_index()

     #7 rename the colum
    .rename(columns={'revenue': 'total_revenue'})
)
print(result)
```
