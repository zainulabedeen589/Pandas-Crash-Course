# MODULE 6 — Merge, Join, Concatenate
# Combining datasets (very common in analytics)

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# -----------------------------
# Sample Data
# -----------------------------
customers = pd.DataFrame({
    'customer_id': [101,102,103,104,105,106],
    'name': ['Alice','Bob','Carol','David','Eve','Frank'],
    'segment': ['Premium','Standard','Premium','Basic','Standard','Premium']
})

orders = pd.DataFrame({
    'order_id': [1,2,3,4,5,6],
    'customer_id': [101,102,101,103,104,999],  # 999 = no match
    'product': ['A','B','A','C','D','A'],
    'revenue': [100,200,150,300,250,400]
})

products = pd.DataFrame({
    'product': ['A','B','C','D'],
    'category': ['Tech','Tech','Home','Home']
})
```

```python
# -----------------------------
# 6.1 Inner Join
# -----------------------------
inner = pd.merge(orders, customers, on='customer_id', how='inner')
print(inner)

```

```python
# -----------------------------
# 6.2 Left Join
# -----------------------------
left = pd.merge(orders, customers, on='customer_id', how='left') # keeps all row from left table (orders) and fills missing customer info with NaN
print(left)
```

```python

# -----------------------------
# 6.3 Anti Join (no match)
# -----------------------------

no_customers = left[left['name'].isna()] # WE find rows that did not match with customers table (customer_id 999)
print(no_customers)
```

```python
# -----------------------------
# 6.4 Multi-table Join
# -----------------------------
full = orders.merge(customers, on='customer_id', how='left') \
             .merge(products, on='product', how='left')

full['margin'] = full['revenue'] * 0.2
print(full)
```

```python
# -----------------------------
# 6.5 Concat (stacking)
# -----------------------------
df1 = pd.DataFrame({'month':['Jan','Jan'],'rev':[100,200]}) # jan data
df2 = pd.DataFrame({'month':['Feb','Feb'],'rev':[300,400]}) # feb data

stacked = pd.concat([df1, df2], ignore_index=True) # stack jan and feb into single table, reset index
print(stacked)
```
