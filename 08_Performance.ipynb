# MODULE 8 — Performance & Large Data
# Optimize memory + speed

```python
import pandas as pd
import numpy as np
```

```python
# -----------------------------
# Large dataset
# -----------------------------
N = 200000

df = pd.DataFrame({
    'id': range(N),
    'category': np.random.choice(['A','B','C'], N),
    'value': np.random.rand(N) * 1000,
    'flag': np.random.choice([0,1], N)
})
```

```python
# -----------------------------
# 8.1 Memory usage
# -----------------------------
print(df.memory_usage(deep=True).sum() / 1024**2, "MB")
```

```python
# -----------------------------
# 8.2 Optimize dtypes
# -----------------------------
df['category'] = df['category'].astype('category')
df['flag'] = df['flag'].astype('int8')
df['value'] = df['value'].astype('float32')

print(df.memory_usage(deep=True).sum() / 1024**2, "MB after")
```

```python
# -----------------------------
# 8.3 Vectorization (FAST)
# -----------------------------
df['new_value'] = np.where(df['value'] > 500,
                df['value'] * 0.9,
                 df['value'])
```

```python
# -----------------------------
# 8.4 Groupby efficiency
# -----------------------------
summary = df.groupby('category')['value'].agg(['mean','sum','count'])
# for each category(A,B,C)
# MEAN -> AVG VAlue
# sum - > total value
# count -> number of rows 
print(summary)
```

```python
# -----------------------------
# 8.5 Avoid loops (important rule)
# -----------------------------
# NEVER use Python loops on big data
df['flagged'] = (df['value'] > 700).astype(int)
# if value > 700 -> 1 else 0
# this run much faster than looping through each row and checking the condition
```
