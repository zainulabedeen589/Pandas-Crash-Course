# MODULE 7 — Time Series Handling
# Used for trends, forecasting, KPIs

```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python

# -----------------------------
# Create time series data
# -----------------------------
dates = pd.date_range('2024-01-01', periods=200)

ts = pd.DataFrame({
    'date': dates,
    'revenue': np.random.randint(1000, 5000, 200),
    'orders': np.random.randint(50, 200, 200)
})

ts.set_index('date', inplace=True)
```

```python
# -----------------------------
# 7.1 Resampling
# -----------------------------
weekly = ts.resample('W').sum() # weekly aggreagation (sum revenue and orders for each week)
monthly = ts.resample('ME').sum() # monthly aggregation (sum revenue and orders for each month, 'ME' means month end)

#print(weekly.head())
print(monthly.head())
```

```python
# -----------------------------
# 7.2 Rolling average
# -----------------------------
ts['7d_avg'] = ts['revenue'].rolling(7).mean() # 7-day rolling average of revenue (average revenue of current day and previous 6 days)
ts['30d_avg'] = ts['revenue'].rolling(30).mean() # 30-day rolling average of revenue (average revenue of current day and previous 29 days)

print(ts.head())
```

```python

# -----------------------------
# 7.3 MoM change
# -----------------------------
# Calculate month-over-month percentage change in revenue using monthly resampled data
monthly_rev = ts['revenue'].resample('ME').sum()

mom = monthly_rev.pct_change()
print(mom)
```

```python
# -----------------------------
# 7.4 Date features
# -----------------------------
df = ts.reset_index()
df['month'] = df['date'].dt.month # extract month from date (1-12)
df['day'] = df['date'].dt.day # Extract day of month from date (1-31)
df['weekday'] = df['date'].dt.day_name() # Extract weekday name from date (e.g. Monday, Tuesday)

print(df.head())
```
